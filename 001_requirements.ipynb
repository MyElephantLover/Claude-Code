```python
## create an API client

from anthropic import AnthropicVertex

client=AnthropicVertex(region="global",project_id="langchain-492000")
model="claude-sonnet-4@20250514"
```

```python
## make a request

message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[{
        "role":"user",
        "content":"what is quantum computing? answer in one setence"
    }
    ]
)
```
